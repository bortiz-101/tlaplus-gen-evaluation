## Parsing into AST

```python
import json
import javalang

def ast_to_dict(node):
    if isinstance(node, set):
        return list(node)
    elif isinstance(node, list):
        return [ast_to_dict(child) for child in node]
    elif isinstance(node, javalang.ast.Node):
        result = {'type': type(node).__name__}
        for attr in node.attrs:
            result[attr] = ast_to_dict(getattr(node, attr))
        return result
    else:
        return node  # basic types (str, int, None, etc.)

java_code = """
public class HelloWorld {
    public static void main(String[] args) {
        System.out.println("Hello, world!");
    }
}
"""

tree = javalang.parse.parse(java_code)
ast_dict = ast_to_dict(tree)

# Save as JSON
with open('ast.ast', 'w') as f:
    json.dump(ast_dict, f, indent=2)
```

```python
import os
import json
import javalang

def ast_to_dict(node):
    """
    Recursively convert javalang AST nodes to a JSON-serializable dict.
    Specifically handle 'set' fields by converting them to lists.
    """
    if isinstance(node, set):
        # JSON doesn't support sets, convert to list
        return list(node)
    elif isinstance(node, list):
        return [ast_to_dict(child) for child in node]
    elif isinstance(node, javalang.ast.Node):
        result = {'type': type(node).__name__}
        for attr in node.attrs:
            result[attr] = ast_to_dict(getattr(node, attr))
        return result
    else:
        # basic types (str, int, float, bool, None)
        return node

if __name__ == "__main__":
    
    from datasets import load_dataset
    original_ds = load_dataset("THUDM/humaneval-x", "java")

    # Directory where .ast files will be stored
    output_dir = "AST"

    # Create the directory if it doesn't already exist
    os.makedirs(output_dir, exist_ok=True)

    # Loop through each snippet in the list
    for row in original_ds['test']:

        print(f"Working on {row['task_id']}")
        if(row['task_id'] == "Java/151"):
            continue
        # Parse Java code into an AST
        tree = javalang.parse.parse(row['test'])

        # Convert the AST to a JSON-serializable dictionary
        ast_dict = ast_to_dict(tree)

        # Construct the file path
        file_name = f"{row['task_id']}.ast"
        file_path = os.path.join(output_dir, file_name)

        # Write the dictionary to the file as JSON
        with open(file_path, 'w') as f:
            json.dump(ast_dict, f, indent=2)

    print("All AST files saved successfully!")
```

#### Revamped remove_block_comments to Save Comment Also

```python
import re

def separate_comments_and_code(java_code):
    """
    Removes all block comments (/* ... */) from the given Java code,
    returning both the cleaned code and the extracted comments.

    Args:
        java_code (str): The Java code as a string.
    
    Returns:
        tuple[str, list[str]]:
            - cleaned_code: The Java code with block comments removed.
            - removed_comments: A list of all removed comments.
    """
    # Regex pattern to match block comments
    pattern = re.compile(r'/\*.*?\*/', re.DOTALL)  # re.DOTALL makes .* match newlines
    
    # Find all block comments
    removed_comments = pattern.findall(java_code)
    
    # Replace block comments with an empty string
    cleaned_code = pattern.sub('', java_code)
    
    return cleaned_code, removed_comments

if __name__ == "__main__":
    code_with_comments = """
public class Example {
    /* This is a block comment */
    public static void main(String[] args) {
        /* Another comment */
        System.out.println("Hello");
        /*
          Multi-line
          comment
        */
    }
}
"""
    cleaned, comments = separate_comments_and_code(code_with_comments)

    print(cleaned)
    print(comments)
```

## Creating Dataset CSV

```python
import os
import csv

def combine_directories_into_csv(
    instructions_dir="instructions/Java",
    sources_dir="sources/Java",
    targets_dir="targets/Java",
    ast_dir="AST/Java",
    output_csv="dataset.csv"
):
    """
    Creates a CSV with four columns:
      - instructions: (string) the content from the .txt file
      - source_path: (string) the relative path to the matching .java in sources/java
      - target_path: (string) the relative path to the matching .java in targets/java
      - ast_path:    (string) the relative path to the matching .ast in AST/java

    The function sorts .txt files by numeric order based on filename.
    E.g., '1.txt', '2.txt', '10.txt' => 1, 2, 10.
    """

    # Prepare list to store CSV rows
    csv_rows = []

    # Check that each directory exists
    for d in [instructions_dir, sources_dir, targets_dir, ast_dir]:
        if not os.path.isdir(d):
            raise FileNotFoundError(f"Directory '{d}' does not exist.")

    # 1) Get all .txt files from instructions/java/
    txt_files = [f for f in os.listdir(instructions_dir) if f.endswith(".txt")]

    # 2) Sort these files by their numeric value rather than lexicographically
    #    This will ensure that '2.txt' < '10.txt', etc.
    def numeric_key(filename):
        # Extract the integer portion from, e.g., '1.txt' => 1
        name_without_ext = os.path.splitext(filename)[0]  # "1", "10", etc.
        return int(name_without_ext)

    txt_files.sort(key=numeric_key)

    # 3) Iterate through the sorted .txt files
    for txt_file in txt_files:
        # Read instructions text
        txt_path = os.path.join(instructions_dir, txt_file)
        with open(txt_path, "r", encoding="utf-8") as f:
            instructions_content = f.read().strip()

        # Derive base name (remove .txt)
        base_name = os.path.splitext(txt_file)[0]  # e.g. '1' from '1.txt'

        # Construct potential file names
        source_candidate = f"{base_name}.java"
        target_candidate = f"{base_name}.java"
        ast_candidate = f"{base_name}.ast"

        # Build full paths and check existence
        source_path = os.path.join(sources_dir, source_candidate)
        if not os.path.isfile(source_path):
            source_path = ""

        target_path = os.path.join(targets_dir, target_candidate)
        if not os.path.isfile(target_path):
            target_path = ""

        ast_path = os.path.join(ast_dir, ast_candidate)
        if not os.path.isfile(ast_path):
            ast_path = ""

        # Append row to our CSV rows
        csv_rows.append([
            instructions_content,
            source_path,
            target_path,
            ast_path
        ])

    # 4) Write out the CSV
    with open(output_csv, mode="w", encoding="utf-8", newline="") as csvfile:
        writer = csv.writer(csvfile)
        # Header
        writer.writerow(["instructions", "source_path", "target_path", "ast_path"])
        # Rows
        writer.writerows(csv_rows)

    print(f"CSV file created: {output_csv}")


if __name__ == "__main__":
    combine_directories_into_csv()
```
